## 7-analyze whether daily change in closing price of stocks or daily returns stocks are co-related or not

```python
closing_price
```

```python
closing_price.columns
```

```python
for col in closing_price.columns:
    closing_price[col + 'pct_chang'] = (closing_price[col]- closing_price[col].shift(1))/closing_price[col].shift(1) * 100
```

```python
closing_price
```

```python
closing_price.columns
```

```python
closing_p = closing_price[['apple_closepct_chang', 'amzn_closepct_chang', 'goog_closepct_chang','msft_closepct_chang']]
```

```python
closing_p
```

```python
g = sns.PairGrid(data = closing_p)
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```
